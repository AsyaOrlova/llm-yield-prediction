```python
uspto_r = pd.read_csv('USPTO_R.csv')
y = uspto_r.yields.tolist()
```

```python
def visualize_embeddings(emb, yields):
  # compute UMAP embeddings
  m_emb = umap.UMAP(metric='cosine', n_jobs=-1, n_components=3).fit_transform(emb)

  # plot reduced embeddings
  fig = plt.figure()
  ax = plt.axes(projection ='3d')
  x = m_emb[:, 0]
  y = m_emb[:, 1]
  z = m_emb[:, 2]
  sc = ax.scatter(x, y, z, c=yields, s=3)
  ax.set(xlabel='UMAP_1', ylabel='UMAP_2', zlabel='UMAP_3')
  cb = plt.colorbar(sc, location = 'right', pad=0.1)
  cb.set_label('Yield, %', labelpad=-1)
  ax.zaxis.labelpad=-1
  plt.tight_layout()
  plt.show()
```

```python
visualize_embeddings(ybert, y)
```

```python
visualize_embeddings(egret, y)
```

```python
visualize_embeddings(chemllm, y)
```

```python
visualize_embeddings(chemdfm, y)
```

```python
visualize_embeddings(gal, y)
```

```python
visualize_embeddings(t5, y)
```

```python
visualize_embeddings(openai, y)
```

```python
visualize_embeddings(llama_text, y)
```

```python
visualize_embeddings(mistral, y)
```
