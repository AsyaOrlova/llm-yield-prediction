```python
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('USPTO_R_text_clf.csv')
df_test = df[df.split=='test']
df_test
```

```python
for k in [2, 4, 6, 8, 10]:
    for seed in [36, 42, 84, 200, 12345]:
        with open(f'result_shots{k}_seed{seed}.txt', 'r') as f:
            lines = [l.rstrip().split('\t')[-1] for l in f.readlines()]
        df_test[f'{k}_{seed}'] = lines
```

```python
df_test['8_84'].value_counts()
```

```python
cols = df_test.columns[4:]
cols
```

```python
for col in cols:
    df_test[col] = df_test[col].apply(lambda x: x.split(':**')[-1])
```

```python
df_test['10_42'].value_counts()
```

```python
for col in cols:
  for idx in df_test.index:
    c = df_test.loc[idx, 'high_yielding']
    answer = df_test.loc[idx, col]
    # print(answer)
    if 'Not high-yielding' in answer or 'not high-yielding' in answer or ('not' in answer and 'high-yielding' in answer) or 'lower' in answer:
      df_test.loc[idx, col] = 0
    elif 'High-yielding' in answer or 'high-yielding' in answer or 'high yield' in answer:
      df_test.loc[idx, col] = 1
    else:
      print(answer)
      df_test.loc[idx, col] = int(not(c))
```

```python
df_test
```

```python
y_true = df_test.high_yielding.tolist()
accs = []
f1s = []
for col in cols:
    y_pred = df_test[col].tolist()
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    accs.append(acc)
    f1s.append(f1)
    if len(accs) < 5:
        continue
    else:
        # print(accs)
        print(col[:2])
        print(f'Accuracy: {np.round(np.mean(accs),2)}+-{np.round(np.std(accs),2)}\nF1-score: {np.round(np.mean(f1s),2)}+-{np.round(np.std(f1s),2)}')
        accs = []
        f1s = []
```
